```python
#Importando as bibliotecas

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider  # Importando a biblioteca para sliders interativos

# Definindo as funes
def plot_seno(A, B, C, D):
    x = np.linspace(-10, 10, 400)  # Cria um array de 400 pontos igualmente espaados entre -10 e 10
    y = A * np.sin(B * x + C) + D  # Calcula os valores da funo seno
    plt.figure(figsize=(10, 4))  # Define o tamanho da figura
    plt.plot(x, y, label=f'{A}*sin({B}*x + {C}) + {D}')  # Plota a funo seno
    plt.title("Funo Seno")  # Define o ttulo do grfico
    plt.legend()  # Adiciona a legenda ao grfico
    plt.grid(True)  # Adiciona uma grade ao grfico
    plt.show()  # Mostra o grfico

def plot_cosseno(A, B, C, D):
    x = np.linspace(-10, 10, 400)  # Cria um array de 400 pontos igualmente espaados entre -10 e 10
    y = A * np.cos(B * x + C) + D  # Calcula os valores da funo cosseno
    plt.figure(figsize=(10, 4))  # Define o tamanho da figura
    plt.plot(x, y, label=f'{A}*cos({B}*x + {C}) + {D}')  # Plota a funo cosseno
    plt.title("Funo Cosseno")  # Define o ttulo do grfico
    plt.legend()  # Adiciona a legenda ao grfico
    plt.grid(True)  # Adiciona uma grade ao grfico
    plt.show()  # Mostra o grfico

def plot_exponencial(A, B, C, D):
    x = np.linspace(-10, 10, 400)  # Cria um array de 400 pontos igualmente espaados entre -10 e 10
    y = A * np.exp(B * x + C) + D  # Calcula os valores da funo exponencial
    plt.figure(figsize=(10, 4))  # Define o tamanho da figura
    plt.plot(x, y, label=f'{A}*exp({B}*x + {C}) + {D}')  # Plota a funo exponencial
    plt.title("Funo Exponencial")  # Define o ttulo do grfico
    plt.legend()  # Adiciona a legenda ao grfico
    plt.grid(True)  # Adiciona uma grade ao grfico
    plt.show()  # Mostra o grfico

def plot_logaritmo(A, B, C):
    x = np.linspace(0.1, 10, 400)  # Cria um array de 400 pontos igualmente espaados entre 0.1 e 10 para evitar log(0)
    y = A * np.log(x + B) + C  # Calcula os valores da funo logaritmo
    plt.figure(figsize=(10, 4))  # Define o tamanho da figura
    plt.plot(x, y, label=f'{A}*log(x + {B}) + {C}')  # Plota a funo logaritmo
    plt.title("Funo Logaritmo")  # Define o ttulo do grfico
    plt.legend()  # Adiciona a legenda ao grfico
    plt.grid(True)  # Adiciona uma grade ao grfico
    plt.show()  # Mostra o grfico

# Interatividade com sliders
# Cria sliders interativos para a funo seno
interact(plot_seno, A=FloatSlider(min=0, max=10, step=0.1, value=4),
         B=FloatSlider(min=0, max=50, step=1, value=36),
         C=FloatSlider(min=-10, max=10, step=0.1, value=6),
         D=FloatSlider(min=-10, max=10, step=0.1, value=-4))

# Cria sliders interativos para a funo cosseno
interact(plot_cosseno, A=FloatSlider(min=-10, max=0, step=0.1, value=-3),
         B=FloatSlider(min=0, max=1, step=0.01, value=0.33333),
         C=FloatSlider(min=0, max=100, step=1, value=60),
         D=FloatSlider(min=0, max=20, step=0.1, value=12))

# Cria sliders interativos para a funo exponencial
interact(plot_exponencial, A=FloatSlider(min=0, max=1, step=0.01, value=0.23),
         B=FloatSlider(min=0, max=10, step=0.1, value=5),
         C=FloatSlider(min=0, max=20, step=1, value=10),
         D=FloatSlider(min=0, max=10, step=0.1, value=2))

# Cria sliders interativos para a funo logaritmo
interact(plot_logaritmo, A=FloatSlider(min=0, max=10, step=0.1, value=2.45),
         B=FloatSlider(min=0, max=20, step=1, value=10),
         C=FloatSlider(min=-10, max=10, step=0.1, value=-2))
```

Comentrios

Funo Seno:

Variando A: Aumenta ou diminui a amplitude da onda senoidal.
Variando B: Altera a frequncia da onda, aumentando ou diminuindo o nmero de oscilaes por unidade de tempo.
Variando C: Desloca a onda horizontalmente, adiantando ou atrasando o grfico.
Variando D: Desloca a onda verticalmente para cima ou para baixo.

Funo Cosseno:

Variando A: Altera a amplitude da onda cossenoidal.
Variando B: Modifica a frequncia da onda.
Variando C: Ajusta a fase da onda, movendo-a horizontalmente.
Variando D: Muda a posio vertical da onda.

Funo Exponencial:

Variando A: Altera a escala da curva exponencial.
Variando B: Modifica a taxa de crescimento ou decrescimento exponencial.
Variando C: Ajusta o ponto de partida na direo x.
Variando D: Desloca a curva verticalmente.

Funo Logaritmo:

Variando A: Modifica a escala da curva logartmica.
Variando B: Ajusta o deslocamento horizontal da curva.
Variando C: Desloca a curva verticalmente.

A visualizao das mudanas em tempo real nos grficos ajuda a entender como cada parmetro afeta as caractersticas das funes.

```python
#Importando as bibliotes
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo de x
x = np.linspace(-10, 10, 400)  # Cria um array de 400 pontos igualmente espaados entre -10 e 10

# Funo Seno original
A1, B1, C1, D1 = 4, 35, 6, -4  # Parmetros da funo seno original
y1 = A1 * np.sin(B1 * x + C1) + D1  # Calcula os valores da funo seno original

# 1. Aumentar o Ganho em 10x
y_gain = 10 * y1  # Multiplica a funo original por 10 para aumentar o ganho

# 2. Somar por 5
y_sum = y1 + 5  # Adiciona 5 a cada valor da funo original

# 3. Subtrair por pi
y_sub = y1 - np.pi  # Subtrai pi de cada valor da funo original

# 4. Fazer um Shift de 3 posies
shift = 3  # Define o nmero de posies para o deslocamento
y_shift = np.roll(y1, shift)  # Desloca a funo original em 3 posies

# Plotando os grficos
fig, axs = plt.subplots(5, 1, figsize=(10, 20))  # Cria um layout de subplots com 5 grficos
fig.suptitle("Transformaes da Funo Seno")  # Ttulo principal do conjunto de grficos

# Grfico da funo original
axs[0].plot(x, y1, label='Funo Original: 4*sin(35*x + 6) - 4', color='blue')
axs[0].set_title("Funo Original")  # Define o ttulo do primeiro grfico
axs[0].legend()  # Adiciona uma legenda ao primeiro grfico

# Grfico do ganho aumentado em 10x
axs[1].plot(x, y_gain, label='Aumentar Ganho em 10x', color='green')
axs[1].set_title("Aumentar Ganho em 10x")  # Define o ttulo do segundo grfico
axs[1].legend()  # Adiciona uma legenda ao segundo grfico

# Grfico da soma por 5
axs[2].plot(x, y_sum, label='Somar por 5', color='red')
axs[2].set_title("Somar por 5")  # Define o ttulo do terceiro grfico
axs[2].legend()  # Adiciona uma legenda ao terceiro grfico

# Grfico da subtrao por pi
axs[3].plot(x, y_sub, label='Subtrair por pi', color='purple')
axs[3].set_title("Subtrair por pi")  # Define o ttulo do quarto grfico
axs[3].legend()  # Adiciona uma legenda ao quarto grfico

# Grfico do shift de 3 posies
axs[4].plot(x, y_shift, label='Shift de 3 posies', color='orange')
axs[4].set_title("Shift de 3 posies")  # Define o ttulo do quinto grfico
axs[4].legend()  # Adiciona uma legenda ao quinto grfico

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Ajusta o layout para que os grficos no se sobreponham
plt.show()  # Mostra todos os grficos
```

```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo de x
x = np.linspace(-9.9, 10, 400)  # Ajustado para evitar log(0) ou log de um nmero negativo

# Funo Seno (1)
A1, B1, C1, D1 = 4, 36, 6, -4  # Parmetros da funo seno
y1 = A1 * np.sin(B1 * x + C1) + D1  # Calcula os valores da funo seno

# Funo Exponencial (3)
A3, B3, C3, D3 = 0.23, 5, 10, 2  # Parmetros da funo exponencial
y3 = A3 * np.exp(x * B3 + C3) + D3  # Calcula os valores da funo exponencial

# Combinao 1 e 3
y_comb1 = y1 + y3  # Soma dos valores das funes seno e exponencial

# Funo Cosseno (2)
A2, B2, C2, D2 = -3, 0.33333, 60, 12  # Parmetros da funo cosseno
y2 = A2 * np.cos(B2 * x + C2) + D2  # Calcula os valores da funo cosseno

# Funo Logaritmo (4)
A4, B4, C4 = 2.45, 10, -2  # Parmetros da funo logaritmo
y4 = A4 * np.log(x + B4) + C4  # Calcula os valores da funo logaritmo

# Combinao 2 e 4
y_comb2 = y2 + y4  # Soma dos valores das funes cosseno e logaritmo

# Aplicando o filtro da mdia com kernel de 3 pontos
kernel_size = 3  # Tamanho do kernel
kernel = np.ones(kernel_size) / kernel_size  # Cria um kernel de mdia de 3 pontos

y_comb1_filtered = np.convolve(y_comb1, kernel, mode='same')  # Aplica o filtro de mdia nos valores combinados de seno e exponencial
y_comb2_filtered = np.convolve(y_comb2, kernel, mode='same')  # Aplica o filtro de mdia nos valores combinados de cosseno e logaritmo

# Calculando mdia, desvio padro e varincia dos sinais filtrados
mean_comb1_filtered = np.mean(y_comb1_filtered)  # Calcula a mdia dos valores filtrados
std_comb1_filtered = np.std(y_comb1_filtered)  # Calcula o desvio padro dos valores filtrados
var_comb1_filtered = np.var(y_comb1_filtered)  # Calcula a varincia dos valores filtrados

mean_comb2_filtered = np.mean(y_comb2_filtered)  # Calcula a mdia dos valores filtrados
std_comb2_filtered = np.std(y_comb2_filtered)  # Calcula o desvio padro dos valores filtrados
var_comb2_filtered = np.var(y_comb2_filtered)  # Calcula a varincia dos valores filtrados

# Plotando os grficos
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # Cria um layout de subplots com 4 grficos
fig.suptitle("Comparao dos Sinais Originais e Filtrados")  # Ttulo principal do conjunto de grficos

# Grficos para a combinao 1 e 3
axs[0, 0].plot(x, y_comb1, label='Original', color='blue')
axs[0, 0].set_title("Combinao Seno e Exponencial - Original")  # Define o ttulo do grfico
axs[0, 1].plot(x, y_comb1_filtered, label='Filtrado', color='cyan')
axs[0, 1].set_title("Combinao Seno e Exponencial - Filtrado")  # Define o ttulo do grfico

# Grficos para a combinao 2 e 4
axs[1, 0].plot(x, y_comb2, label='Original', color='green')
axs[1, 0].set_title("Combinao Cosseno e Logaritmo - Original")  # Define o ttulo do grfico
axs[1, 1].plot(x, y_comb2_filtered, label='Filtrado', color='lime')
axs[1, 1].set_title("Combinao Cosseno e Logaritmo - Filtrado")  # Define o ttulo do grfico

for ax in axs.flat:
    ax.legend()  # Adiciona uma legenda a cada grfico

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Ajusta o layout para que os grficos no se sobreponham
plt.show()  # Mostra todos os grficos

# Resultados
print("Combinao Seno e Exponencial - Original")
print(f"Mdia: {mean_comb1:.2f}, Desvio Padro: {std_comb1:.2f}, Varincia: {var_comb1:.2f}")
print("Combinao Seno e Exponencial - Filtrado")
print(f"Mdia: {mean_comb1_filtered:.2f}, Desvio Padro: {std_comb1_filtered:.2f}, Varincia: {var_comb1_filtered:.2f}\n")

print("Combinao Cosseno e Logaritmo - Original")
print(f"Mdia: {mean_comb2:.2f}, Desvio Padro: {std_comb2:.2f}, Varincia: {var_comb2:.2f}")
print("Combinao Cosseno e Logaritmo - Filtrado")
print(f"Mdia: {mean_comb2_filtered:.2f}, Desvio Padro: {std_comb2_filtered:.2f}, Varincia: {var_comb2_filtered:.2f}")
```
