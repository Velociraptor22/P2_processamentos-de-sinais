```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo de x
x = np.linspace(-9.99, 10, 400)  # Esse ajuste é necessário para evitar log(0) ou log de um número negativo

# Função Seno
A1, B1, C1, D1 = 4, 36, 6, -4
y1 = A1 * np.sin(B1 * x + C1) + D1
# A função seno é dada por 4*sin(36*x + 6) - 4.
# O gráfico da função seno oscila entre -4 e 4 com uma amplitude multiplicada por 4.

# Função Cosseno
A2, B2, C2, D2 = -3, 0.33333, 60, 12
y2 = A2 * np.cos(B2 * x + C2) + D2
# A função cosseno é dada por -3*cos(0.33333*x + 60) + 12.
# O gráfico da função cosseno oscila entre 9 e 15 com uma amplitude multiplicada por -3.

# Função Exponencial
A3, B3, C3, D3 = 0.23, 5, 10, 2
y3 = A3 * np.exp(x * B3 + C3) + D3
# A função exponencial é dada por 0.23*exp(5*x + 10) + 2.
# O gráfico da função exponencial cresce rapidamente devido ao fator exponencial.

# Função Logaritmo
A4, B4, C4 = 2.45, 10, -2
y4 = A4 * np.log(x + B4) + C4
# A função logaritmo é dada por 2.45*log(x + 10) - 2.
# O gráfico da função logarítmica começa em -2 e cresce logarit1macamente.

# Plotando os gráficos
fig, axs = plt.subplots(4, 1, figsize=(10, 20))
fig.suptitle("Gráficos das Funções Especificadas")

axs[0].plot(x, y1, label='4*sin(36*x + 6) - 4', color='blue')
axs[0].set_title("Função Seno")
axs[0].legend()

axs[1].plot(x, y2, label='-3*cos(0.33333*x + 60) + 12', color='green')
axs[1].set_title("Função Cosseno")
axs[1].legend()

axs[2].plot(x, y3, label='0.23*exp(5*x + 10) + 2', color='red')
axs[2].set_title("Função Exponencial")
axs[2].legend()

axs[3].plot(x, y4, label='2.45*log(x + 10) - 2', color='purple')
axs[3].set_title("Função Logaritmo")
axs[3].legend()

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

#Exercício 1.1
Faça variação dos parametros A,B,C e D do exercício anterior e comente os resultados Dica: utilize um "slide bar" para fazer a variação

```python
#Importando as bibliotecas

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider  # Importando a biblioteca para sliders interativos

# Definindo as funções
def plot_seno(A, B, C, D):
    x = np.linspace(-10, 10, 400)  # Cria um array de 400 pontos igualmente espaçados entre -10 e 10
    y = A * np.sin(B * x + C) + D  # Calcula os valores da função seno
    plt.figure(figsize=(10, 4))  # Define o tamanho da figura
    plt.plot(x, y, label=f'{A}*sin({B}*x + {C}) + {D}')  # Plota a função seno
    plt.title("Função Seno")  # Define o título do gráfico
    plt.legend()  # Adiciona a legenda ao gráfico
    plt.grid(True)  # Adiciona uma grade ao gráfico
    plt.show()  # Mostra o gráfico

def plot_cosseno(A, B, C, D):
    x = np.linspace(-10, 10, 400)  # Cria um array de 400 pontos igualmente espaçados entre -10 e 10
    y = A * np.cos(B * x + C) + D  # Calcula os valores da função cosseno
    plt.figure(figsize=(10, 4))  # Define o tamanho da figura
    plt.plot(x, y, label=f'{A}*cos({B}*x + {C}) + {D}')  # Plota a função cosseno
    plt.title("Função Cosseno")  # Define o título do gráfico
    plt.legend()  # Adiciona a legenda ao gráfico
    plt.grid(True)  # Adiciona uma grade ao gráfico
    plt.show()  # Mostra o gráfico

def plot_exponencial(A, B, C, D):
    x = np.linspace(-10, 10, 400)  # Cria um array de 400 pontos igualmente espaçados entre -10 e 10
    y = A * np.exp(B * x + C) + D  # Calcula os valores da função exponencial
    plt.figure(figsize=(10, 4))  # Define o tamanho da figura
    plt.plot(x, y, label=f'{A}*exp({B}*x + {C}) + {D}')  # Plota a função exponencial
    plt.title("Função Exponencial")  # Define o título do gráfico
    plt.legend()  # Adiciona a legenda ao gráfico
    plt.grid(True)  # Adiciona uma grade ao gráfico
    plt.show()  # Mostra o gráfico

def plot_logaritmo(A, B, C):
    x = np.linspace(0.1, 10, 400)  # Cria um array de 400 pontos igualmente espaçados entre 0.1 e 10 para evitar log(0)
    y = A * np.log(x + B) + C  # Calcula os valores da função logaritmo
    plt.figure(figsize=(10, 4))  # Define o tamanho da figura
    plt.plot(x, y, label=f'{A}*log(x + {B}) + {C}')  # Plota a função logaritmo
    plt.title("Função Logaritmo")  # Define o título do gráfico
    plt.legend()  # Adiciona a legenda ao gráfico
    plt.grid(True)  # Adiciona uma grade ao gráfico
    plt.show()  # Mostra o gráfico

# Interatividade com sliders
# Cria sliders interativos para a função seno
interact(plot_seno, A=FloatSlider(min=0, max=10, step=0.1, value=4),
         B=FloatSlider(min=0, max=50, step=1, value=36),
         C=FloatSlider(min=-10, max=10, step=0.1, value=6),
         D=FloatSlider(min=-10, max=10, step=0.1, value=-4))

# Cria sliders interativos para a função cosseno
interact(plot_cosseno, A=FloatSlider(min=-10, max=0, step=0.1, value=-3),
         B=FloatSlider(min=0, max=1, step=0.01, value=0.33333),
         C=FloatSlider(min=0, max=100, step=1, value=60),
         D=FloatSlider(min=0, max=20, step=0.1, value=12))

# Cria sliders interativos para a função exponencial
interact(plot_exponencial, A=FloatSlider(min=0, max=1, step=0.01, value=0.23),
         B=FloatSlider(min=0, max=10, step=0.1, value=5),
         C=FloatSlider(min=0, max=20, step=1, value=10),
         D=FloatSlider(min=0, max=10, step=0.1, value=2))

# Cria sliders interativos para a função logaritmo
interact(plot_logaritmo, A=FloatSlider(min=0, max=10, step=0.1, value=2.45),
         B=FloatSlider(min=0, max=20, step=1, value=10),
         C=FloatSlider(min=-10, max=10, step=0.1, value=-2))
```

Comentários

Função Seno:

Variando A: Aumenta ou diminui a amplitude da onda senoidal.
Variando B: Altera a frequência da onda, aumentando ou diminuindo o número de oscilações por unidade de tempo.
Variando C: Desloca a onda horizontalmente, adiantando ou atrasando o gráfico.
Variando D: Desloca a onda verticalmente para cima ou para baixo.

Função Cosseno:

Variando A: Altera a amplitude da onda cossenoidal.
Variando B: Modifica a frequência da onda.
Variando C: Ajusta a fase da onda, movendo-a horizontalmente.
Variando D: Muda a posição vertical da onda.

Função Exponencial:

Variando A: Altera a escala da curva exponencial.
Variando B: Modifica a taxa de crescimento ou decrescimento exponencial.
Variando C: Ajusta o ponto de partida na direção x.
Variando D: Desloca a curva verticalmente.

Função Logaritmo:

Variando A: Modifica a escala da curva logarítmica.
Variando B: Ajusta o deslocamento horizontal da curva.
Variando C: Desloca a curva verticalmente.

A visualização das mudanças em tempo real nos gráficos ajuda a entender como cada parâmetro afeta as características das funções.

```python
#Importando as bibliotes
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo de x
x = np.linspace(-10, 10, 400)  # Cria um array de 400 pontos igualmente espaçados entre -10 e 10

# Função Seno original
A1, B1, C1, D1 = 4, 35, 6, -4  # Parâmetros da função seno original
y1 = A1 * np.sin(B1 * x + C1) + D1  # Calcula os valores da função seno original

# 1. Aumentar o Ganho em 10x
y_gain = 10 * y1  # Multiplica a função original por 10 para aumentar o ganho

# 2. Somar por 5
y_sum = y1 + 5  # Adiciona 5 a cada valor da função original

# 3. Subtrair por pi
y_sub = y1 - np.pi  # Subtrai pi de cada valor da função original

# 4. Fazer um Shift de 3 posições
shift = 3  # Define o número de posições para o deslocamento
y_shift = np.roll(y1, shift)  # Desloca a função original em 3 posições

# Plotando os gráficos
fig, axs = plt.subplots(5, 1, figsize=(10, 20))  # Cria um layout de subplots com 5 gráficos
fig.suptitle("Transformações da Função Seno")  # Título principal do conjunto de gráficos

# Gráfico da função original
axs[0].plot(x, y1, label='Função Original: 4*sin(35*x + 6) - 4', color='blue')
axs[0].set_title("Função Original")  # Define o título do primeiro gráfico
axs[0].legend()  # Adiciona uma legenda ao primeiro gráfico

# Gráfico do ganho aumentado em 10x
axs[1].plot(x, y_gain, label='Aumentar Ganho em 10x', color='green')
axs[1].set_title("Aumentar Ganho em 10x")  # Define o título do segundo gráfico
axs[1].legend()  # Adiciona uma legenda ao segundo gráfico

# Gráfico da soma por 5
axs[2].plot(x, y_sum, label='Somar por 5', color='red')
axs[2].set_title("Somar por 5")  # Define o título do terceiro gráfico
axs[2].legend()  # Adiciona uma legenda ao terceiro gráfico

# Gráfico da subtração por pi
axs[3].plot(x, y_sub, label='Subtrair por pi', color='purple')
axs[3].set_title("Subtrair por pi")  # Define o título do quarto gráfico
axs[3].legend()  # Adiciona uma legenda ao quarto gráfico

# Gráfico do shift de 3 posições
axs[4].plot(x, y_shift, label='Shift de 3 posições', color='orange')
axs[4].set_title("Shift de 3 posições")  # Define o título do quinto gráfico
axs[4].legend()  # Adiciona uma legenda ao quinto gráfico

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Ajusta o layout para que os gráficos não se sobreponham
plt.show()  # Mostra todos os gráficos
```

#Exercicio 3
Combine as seguintes funções do exercicio 1:

Tire a média, o desvio padrão e a variância. Comente os resultados



```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo de x
x = np.linspace(-9.9, 10, 400)  # Ajustado para evitar log(0) ou log de um número negativo

# Função Seno (1)
A1, B1, C1, D1 = 4, 36, 6, -4  # Parâmetros da função seno
y1 = A1 * np.sin(B1 * x + C1) + D1  # Calcula os valores da função seno

# Função Exponencial (3)
A3, B3, C3, D3 = 0.23, 5, 10, 2  # Parâmetros da função exponencial
y3 = A3 * np.exp(x * B3 + C3) + D3  # Calcula os valores da função exponencial

# Combinação 1 e 3
y_comb1 = y1 + y3  # Soma dos valores das funções seno e exponencial

# Função Cosseno (2)
A2, B2, C2, D2 = -3, 0.33333, 60, 12  # Parâmetros da função cosseno
y2 = A2 * np.cos(B2 * x + C2) + D2  # Calcula os valores da função cosseno

# Função Logaritmo (4)
A4, B4, C4 = 2.45, 10, -2  # Parâmetros da função logaritmo
y4 = A4 * np.log(x + B4) + C4  # Calcula os valores da função logaritmo

# Combinação 2 e 4
y_comb2 = y2 + y4  # Soma dos valores das funções cosseno e logaritmo

# Aplicando o filtro da média com kernel de 3 pontos
kernel_size = 3  # Tamanho do kernel
kernel = np.ones(kernel_size) / kernel_size  # Cria um kernel de média de 3 pontos

y_comb1_filtered = np.convolve(y_comb1, kernel, mode='same')  # Aplica o filtro de média nos valores combinados de seno e exponencial
y_comb2_filtered = np.convolve(y_comb2, kernel, mode='same')  # Aplica o filtro de média nos valores combinados de cosseno e logaritmo

# Calculando média, desvio padrão e variância dos sinais filtrados
mean_comb1_filtered = np.mean(y_comb1_filtered)  # Calcula a média dos valores filtrados
std_comb1_filtered = np.std(y_comb1_filtered)  # Calcula o desvio padrão dos valores filtrados
var_comb1_filtered = np.var(y_comb1_filtered)  # Calcula a variância dos valores filtrados

mean_comb2_filtered = np.mean(y_comb2_filtered)  # Calcula a média dos valores filtrados
std_comb2_filtered = np.std(y_comb2_filtered)  # Calcula o desvio padrão dos valores filtrados
var_comb2_filtered = np.var(y_comb2_filtered)  # Calcula a variância dos valores filtrados

# Plotando os gráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # Cria um layout de subplots com 4 gráficos
fig.suptitle("Comparação dos Sinais Originais e Filtrados")  # Título principal do conjunto de gráficos

# Gráficos para a combinação 1 e 3
axs[0, 0].plot(x, y_comb1, label='Original', color='blue')
axs[0, 0].set_title("Combinação Seno e Exponencial - Original")  # Define o título do gráfico
axs[0, 1].plot(x, y_comb1_filtered, label='Filtrado', color='cyan')
axs[0, 1].set_title("Combinação Seno e Exponencial - Filtrado")  # Define o título do gráfico

# Gráficos para a combinação 2 e 4
axs[1, 0].plot(x, y_comb2, label='Original', color='green')
axs[1, 0].set_title("Combinação Cosseno e Logaritmo - Original")  # Define o título do gráfico
axs[1, 1].plot(x, y_comb2_filtered, label='Filtrado', color='lime')
axs[1, 1].set_title("Combinação Cosseno e Logaritmo - Filtrado")  # Define o título do gráfico

for ax in axs.flat:
    ax.legend()  # Adiciona uma legenda a cada gráfico

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Ajusta o layout para que os gráficos não se sobreponham
plt.show()  # Mostra todos os gráficos

# Resultados
print("Combinação Seno e Exponencial - Original")
print(f"Média: {mean_comb1:.2f}, Desvio Padrão: {std_comb1:.2f}, Variância: {var_comb1:.2f}")
print("Combinação Seno e Exponencial - Filtrado")
print(f"Média: {mean_comb1_filtered:.2f}, Desvio Padrão: {std_comb1_filtered:.2f}, Variância: {var_comb1_filtered:.2f}\n")

print("Combinação Cosseno e Logaritmo - Original")
print(f"Média: {mean_comb2:.2f}, Desvio Padrão: {std_comb2:.2f}, Variância: {var_comb2:.2f}")
print("Combinação Cosseno e Logaritmo - Filtrado")
print(f"Média: {mean_comb2_filtered:.2f}, Desvio Padrão: {std_comb2_filtered:.2f}, Variância: {var_comb2_filtered:.2f}")
```
